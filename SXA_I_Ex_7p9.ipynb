```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Conversion Range (Xa)
Xa = np.arange(0.01, 1.0, 0.01)

# Model Parameters
tau = 5             # Residence time (min)
rho = 0.75          # Average Density (g/cm^3)
CAo = 0.002         # Inlet Stream Concentration (mol/cm^3)
CpA = 35            # Average Heat Capacity (cal/g/K)
k0 = 5e10           # Frequency coefficient (1/min)
E = 7000            # Activation Energy /R (cm^3/mol/min)
DHr = 40000         # Heat of Reaction (cal/kmol)
To = np.arange(280, 330, 10)  # Inlet Stream Temperature (K)

# Mass and Energy Balances
T_mb = 7000 / (np.log(10e8) + 2 * np.log(1 - Xa) - np.log(Xa))  # Temperature from mass balance

# Energy balance calculations for each To
T_eb = np.zeros((len(Xa), len(To)))
for i in range(len(To)):
    T_eb[:, i] = To[i] + 133.33 * Xa  # Temperature from energy balance for each To

# Plotting
plt.figure(dpi=200)
cmap = plt.get_cmap('jet', len(To))

# Plot T_mb
plt.plot(Xa, T_mb, 'k-', linewidth=3, label='$T-IM$')

# Plot each T_eb curve in the colormap
for i in range(len(To)):
    plt.plot(Xa, T_eb[:, i], color=cmap(i), linewidth=1.5, label=f'$T-IE$ for To = {To[i]}')

# Add grid, legend, and labels
plt.grid(True)
plt.ylabel('Θερμοκρασία, Τ [R]', fontsize=14)
plt.xlabel('Βαθμός Μετατροπής, $X_A$', fontsize=14)
#plt.title("Θερμοκρασία vs Βαθμός Μετατροπής")
plt.legend(fontsize=10)
plt.show()
```
