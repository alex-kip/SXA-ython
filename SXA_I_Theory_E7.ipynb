```python
# Import necessary libraries
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the system of differential equations
def SXA_I_Dynamical_Systems_Theory(t, x):
    # Model Parameters
    k1 = 0.5        # reaction constant (m^3*mol^-1*h^-1)
    # Differential Equations
    dx = -k1 * x    # d[CA]/dt
    return dx

# Set time span for the integration (e.g., from t=0 to t=10 hours)
t_span = (0, 5)  # time span in hours
t_eval = np.linspace(0, 5, 100)  # time points for evaluation

# Initial concentrations: [CA0]
initial_conditions = [5]  # Initial values for [CA]

# Solve the ODE system using solve_ivp
sol = solve_ivp(SXA_I_Dynamical_Systems_Theory, t_span, initial_conditions, t_eval=t_eval)

# Extract the solution for each component
CA = sol.y[0, :]
time = sol.t

# Plot the results
plt.figure(dpi=150)
plt.plot(time, CA, label='[$C_A$] (mol/$m^3$)', color='blue')

# Ensure plot starts at origin
plt.xlim(left=0)  # X-axis starts at 0
plt.ylim(bottom=0)  # Y-axis starts at 0

plt.xlabel('Χρόνος (hours)', fontsize=12)
plt.ylabel('Συγκέντρωση (mol/$m^3$)', fontsize=12)
#plt.title('Concentration Profiles of Species', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the system of differential equations
def SXA_I_Dynamical_Systems_Theory2(t, n):
    # Model Parameters
    k0 = 2.5e7        # reaction constant (m^3*mol^-1*h^-1)
    E = 12000        # reaction constant (m^3*mol^-1*h^-1)
    R = 1.987        # reaction constant (m^3*mol^-1*h^-1)
    DHrt = -50000        # reaction constant (m^3*mol^-1*h^-1)
    Cp = 0.9        # reaction constant (m^3*mol^-1*h^-1)
    rho = 1.05        # reaction constant (m^3*mol^-1*h^-1)
    V = 150        # reaction constant (m^3*mol^-1*h^-1)
    
    # Rate equation
    R = k0 * np.exp(-E/(R*n[1])) * n[0]
    
    
    # Differential Equations
    dn = np.zeros(2)
    dn[0] = -R    # d[CA]/dt
    dn[1] = -DHrt * R * V / (rho * V * Cp)
    
    return dn

# Set time span for the integration (e.g., from t=0 to t=10 hours)
t_span = (0, 20)  # time span in hours
t_eval = np.linspace(0, 20, 100)  # time points for evaluation

# Initial concentrations: [CA0]
initial_conditions = [0.0009, 300]  # Initial values for [CA]

# Solve the ODE system using solve_ivp
sol = solve_ivp(SXA_I_Dynamical_Systems_Theory2, t_span, initial_conditions, t_eval=t_eval)

# Extract the solution for each component
CA = sol.y[0, :]
T = sol.y[1, :]
time = sol.t

# Plot the results
plt.figure(dpi=150)
plt.plot(time, CA*10000, label='[$C_A$] (mol/$m^3$)', color='blue')

# Ensure plot starts at origin
plt.xlim(left=0)  # X-axis starts at 0
plt.ylim(bottom=0)  # Y-axis starts at 0

plt.xlabel('Χρόνος (hours)', fontsize=12)
plt.ylabel('Συγκέντρωση (mol/$m^3$)', fontsize=12)
#plt.title('Concentration Profiles of Species', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()

# Plot the results
plt.figure(dpi=150)
plt.plot(time, T, label='[$T$] (mol/$m^3$)', color='red')

# Ensure plot starts at origin
plt.xlim(left=0)  # X-axis starts at 0
plt.ylim(bottom=270)  # Y-axis starts at 0

plt.xlabel('Χρόνος (hours)', fontsize=12)
plt.ylabel('Θερμοκρασία (Κ)', fontsize=12)
#plt.title('Concentration Profiles of Species', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
